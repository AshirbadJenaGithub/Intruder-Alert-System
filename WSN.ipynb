```python
import tensorflow as tf


def build_model(hp):
  model = tf.keras.models.Sequential([
      tf.keras.layers.Dense(32, input_shape=(x_train[1].shape[0],1), activation='relu'),
      tf.keras.layers.GRU(units=hp.Int('units', min_value=32, max_value=128), return_sequences=True),
      tf.keras.layers.GRU(units=hp.Int('units', min_value=32, max_value=128), return_sequences=False, activation='tanh'),
      tf.keras.layers.Dense(64, activation='softmax')
  ])

  model.compile(loss="sparse_categorical_crossentropy",
                optimizer=tf.keras.optimizers.Adam(learning_rate=0.02),
                metrics=['accuracy'])

  model.fit(x_train, y_train, epochs=10)
  return model
```

```python
tuner=RandomSearch(build_model,objective='val_accuracy',max_trials=5)
tuner.search(x_train,y_train,epochs=5)
```
